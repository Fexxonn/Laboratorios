```python
# Imprimir un mensaje en la consola
print("¡Hola! Bienvenido a tu primer programa en Python.")

# Declarar variables de diferentes tipos y realizar operaciones matemáticas simples
numero_entero = 10
numero_decimal = 5.5
texto = "Hola, mundo"
resultado_suma = numero_entero + numero_decimal

print(f"El resultado de sumar {numero_entero} y {numero_decimal} es {resultado_suma}")

# Concatenar cadenas de texto y utilizar funciones básicas
nombre = input("¿Cuál es tu nombre? ")
print("Hola " + nombre + ", ¡es un gusto conocerte!")
```

```python
# Determinar si un número es par o impar
numero = int(input("Introduce un número: "))
if numero % 2 == 0:
    print(f"El número {numero} es par.")
else:
    print(f"El número {numero} es impar.")

# Bucle for para imprimir los cuadrados de una lista de números
numeros = [1, 2, 3, 4, 5]
print("Los cuadrados de los números son:")
for n in numeros:
    print(n ** 2)

# Bucle while para solicitar repetidamente hasta cumplir una condición
intento = ""
while intento.lower() != "salir":
    intento = input("Escribe 'salir' para terminar: ")
    if intento.lower() != "salir":
        print("Aún no has salido.")
print("¡Gracias por participar!")
```

```python
# Mostrar los nombres de estudiantes usando un bucle
estudiantes = ["Ana", "Juan", "María", "Carlos"]
print("Lista de estudiantes:")
for estudiante in estudiantes:
    print(estudiante)

# Crear un diccionario simple y mostrar claves y valores
contactos = {"Ana": "ana@example.com", "Juan": "juan@example.com"}
print("Información de contacto:")
for nombre, correo in contactos.items():
    print(f"Nombre: {nombre}, Correo: {correo}")

# Agregar elementos a una lista o actualizar un diccionario
nueva_lista = []
opcion = input("¿Quieres agregar algo a la lista? (sí/no): ")
while opcion.lower() == "sí":
    elemento = input("Escribe el elemento: ")
    nueva_lista.append(elemento)
    opcion = input("¿Quieres agregar más elementos? (sí/no): ")
print("Tu lista final es:", nueva_lista)

# Actualizar diccionario
clave = input("Escribe un nombre para agregar o actualizar en contactos: ")
valor = input("Escribe el correo: ")
contactos[clave] = valor
print("Diccionario actualizado:", contactos)
```

```python
# Calculadora básica
print("Bienvenido a la calculadora.")
operacion = input("Elige una operación (suma, resta, multiplicación, división): ").lower()
num1 = float(input("Introduce el primer número: "))
num2 = float(input("Introduce el segundo número: "))

if operacion == "suma":
    print(f"El resultado es: {num1 + num2}")
elif operacion == "resta":
    print(f"El resultado es: {num1 - num2}")
elif operacion == "multiplicación":
    print(f"El resultado es: {num1 * num2}")
elif operacion == "división":
    if num2 != 0:
        print(f"El resultado es: {num1 / num2}")
    else:
        print("No se puede dividir entre cero.")
else:
    print("Operación no válida.")

# Juego de adivinanza
import random

numero_secreto = random.randint(1, 100)
print("¡Adivina el número entre 1 y 100!")
adivinado = False

while not adivinado:
    intento = int(input("Tu intento: "))
    if intento < numero_secreto:
        print("El número es mayor.")
    elif intento > numero_secreto:
        print("El número es menor.")
    else:
        print("¡Felicidades! Adivinaste el número.")
        adivinado = True
```
